```python
import sys
```

```python
f = open('reviews.txt')
raw_reviews = f.readlines()
f.close()

f = open('labels.txt')
raw_labels = f.readlines()
f.close()

tokens = list(map(lambda x:set(x.split(" ")), raw_reviews))
```

```python
tokens
```

```python
vocab = set()
for sent in tokens:
    for word in sent:
        if(len(word)>0):
            vocab.add(word)
vocab = list(vocab)
```

```python
vocab
```

```python
word2index = {}
for i, word in enumerate(vocab):
    word2index[word]=i
```

```python
word2index
```

```python
input_dataset = list()
for sent in tokens:
    sent_indices = list()
    for word in sent:
        try:
            sent_indices.append(word2index[word])
        except:
            ""
    input_dataset.append(list(set(sent_indices)))

target_dataset = list()
for label in raw_labels:
    if label == 'positive\n':
        target_dataset.append(1)
    else:
        target_dataset.append(0)         
```

```python
# taking vector of 1s and 0s is equiv to summing several rows of a matrix. so instead of doing big vector matrix multiplication
# we can select the relevant rows of weights_0_1 and sum them. b/c sentiment vocab is around 70k words, most of matrix multi
# would be spent multiplying 0s. summing is much more efficient. using the process of selecting rows and performing a sum
# means treating the first linear layer (weights_0_1) as an embedding layer. layer_1 will still be the same, but summing 
# small number of rows is faster.
```

```python
np.random.seed(1)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

alpha, iterations = (0.01, 2)
hidden_size = 100

weights_0_1 = 0.2*np.random.random((len(vocab),hidden_size)) - 0.1
weights_1_2 = 0.2*np.random.random((hidden_size, 1)) - 0.1
```

```python
correct, total = (0, 0)
for iter in range(iterations):
    for i in range(len(input_dataset)-1000): # train on 1st 24,000 reviews
        x, y = (input_dataset[i], target_dataset[i])
        layer_1 = sigmoid(np.sum(weights_0_1[x], axis=0)) # embed + sigmoid
        layer_2 = sigmoid(np.dot(layer_1, weights_1_2)) # linear + softmax
        layer_2_delta = layer_2 - y
        layer_1_delta = layer_2_delta.dot(weights_1_2.T)
        weights_0_1[x] -= layer_1_delta * alpha
        weights_1_2 -= np.outer(layer_1, layer_2_delta) * alpha
    
        if(np.abs(layer_2_delta) < 0.5):
            correct += 1
        total += 1
        if(i % 10 == 9):
            progress = str(i/float(len(input_dataset)))
            sys.stdout.write(
                f'\rIter:{iter} '
                f'progress:{progress[2:4]}.{progress[4:6]}% '
                f'training acc:{correct/float(total):.2f}%'
            )
    print()
```

```python
correct, total=(0, 0)
for i in range(len(input_dataset)-1000,len(input_dataset)):
    x = input_dataset[i]
    y = target_dataset[i]
    
    layer_1 = sigmoid(np.sum(weights_0_1[x],axis=0))
    layer_2 = sigmoid(np.dot(layer_1, weights_1_2))
    
    if(np.abs(layer_2-y) < 0.5):
        correct += 1
    total += 1
print("test acc:" + str(correct / float(total)))
```
